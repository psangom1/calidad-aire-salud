```python


# --- Paquetes ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1) Cargar el CSV ---
# Si el archivo est√° en el mismo directorio del notebook, basta con:
df = pd.read_csv('df_final5.csv')

# --- 2) Seleccionar y limpiar columnas relevantes ---
cols = ['pm25', 'Total', 'ciudad', 'comunidad_norm']
df.columns = [c.strip() for c in df.columns]  # normaliza espacios
df_plot = df[[c for c in cols if c in df.columns]].copy()

# Convierte a num√©rico por si hay strings y gestiona NaN
for c in ['pm25', 'Total']:
    if c in df_plot.columns:
        df_plot[c] = pd.to_numeric(df_plot[c], errors='coerce')

# Quita filas con NaN en pm25 o Total
df_plot = df_plot.dropna(subset=['pm25', 'Total'])

# --- 3) Gr√°fico de dispersi√≥n con l√≠nea de tendencia ---
plt.figure(figsize=(8, 6))
sns.set(style='whitegrid')
sns.regplot(
    data=df_plot,
    x='pm25', y='Total',
    scatter_kws={'alpha': 0.8},
    line_kws={'color': 'red'}
)
plt.title('Relaci√≥n entre PM2.5 y Total')
plt.xlabel('PM2.5 (¬µg/m¬≥)')
plt.ylabel('Total')
plt.tight_layout()
plt.show()

# --- 4) Correlaci√≥n de Pearson ---
corr = df_plot['pm25'].corr(df_plot['Total'])
print(f'Correlaci√≥n de Pearson (pm25 vs Total): {corr:.3f}')
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1) Cargar el CSV ---
df = pd.read_csv('df_final5.csv')

# --- 2) Seleccionar y limpiar columnas relevantes ---
cols = ['no2', 'Total', 'ciudad', 'comunidad_norm']
df.columns = [c.strip() for c in df.columns]  # normaliza espacios
df_plot = df[[c for c in cols if c in df.columns]].copy()

# Convertimos a num√©rico por si hay strings y gestionamos NaN
for c in ['no2', 'Total']:
    if c in df_plot.columns:
        df_plot[c] = pd.to_numeric(df_plot[c], errors='coerce')

# Quitamos filas con NaN en no2 o Total
df_plot = df_plot.dropna(subset=['no2', 'Total'])

# --- 3) Gr√°fico de dispersi√≥n con l√≠nea de tendencia ---
plt.figure(figsize=(8, 6))
sns.set(style='whitegrid')
sns.regplot(
    data=df_plot,
    x='no2', y='Total',
    scatter_kws={'alpha': 0.8},
    line_kws={'color': 'red'}
)
plt.title('Relaci√≥n entre NO‚ÇÇ y Total')
plt.xlabel('NO‚ÇÇ (¬µg/m¬≥)')
plt.ylabel('Total')
plt.tight_layout()
plt.show()

# --- 4) Correlaci√≥n de Pearson ---
corr = df_plot['no2'].corr(df_plot['Total'])
print(f'Correlaci√≥n de Pearson (no2 vs Total): {corr:.3f}')
```
